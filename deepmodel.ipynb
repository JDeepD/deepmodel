#### Tensors are multi-dimensional arrays with a uniform type (called a dtype). You can see all supported dtypes at tf.dtypes.DType. 
###### All tensors are immutable like Python numbers and strings: you can never update the contents of a tensor, only create a new one.


### Rank of tensors :

Def : "The rank of a tensor is analagous to representing a point in a dimension. For example representing a scalar on x-axis is a rank-0 tensor. Representing a vector on X,Y,Z space is a rank 1 tensor. A matrix is a rank-2 tensor." See more here https://mathworld.wolfram.com/TensorRank.html

Rank of a tensor can be found using `tf.rank(name of tensor)`

-------------------

### Shape of tensors
Def : "It is analogous to the order of a matrix i.e Rows x Columns. Since a tensor is an n-dimensional entity, it can have n-shapes. For example :"
The shape can be found using tensorname.shape

```
z = tf.constant(4) # z is a rank 0 tensor and its shape is null.
```
```
x = tf.Variable([2,4,5],tf.int32) # a is a rank 1 tensor (i.e a vector) and its shape is 3  
```

```
i = tf.Variable([[o,p],[q,r]] , tf.int16) # It has a rank of (2,2,5) .Note that this is a 3 dimensional matrix(i.e a tensor) 
```
