```python
from __future__ import print_function ,absolute_import , division , unicode_literals
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output
from six.moves import urllib
import tensorflow.compat.v2.feature_column as fc
import tensorflow as tf

dftrain = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/train.csv')
y_train = dftrain.pop('survived')

dfeval = pd.read_csv("https://storage.googleapis.com/tf-datasets/titanic/eval.csv")
y_eval = dfeval.pop("survived")

cat_cols = ["sex" , "class" , "deck" , "embark_town" , "alone","n_siblings_spouses" ,"parch"]
num_cols = ["age"  ,"fare" ]

fc = []


for i in cat_cols:
    x = dftrain[i].unique()
    fc.append(tf.feature_column.categorical_column_with_vocabulary_list(i , x))

for j in num_cols:
    fc.append(tf.feature_column.numeric_column(j , dtype=tf.float32))


def input_fn(dftrain , dflabel , shuffle=True  , epoch =50, bat_size = 32):
    
    def fn():
        dataset = tf.data.Dataset.from_tensor_slices( (dict(dftrain) , dflabel))
        if shuffle:
            dataset = dataset.shuffle(1000)
        dataset = dataset.batch(bat_size).repeat(epoch)
        return(dataset)
    return(fn)

train_input_fn = input_fn(dftrain, y_train)
eval_input_fn = input_fn(dfeval , y_eval , epoch=1 ,shuffle=False)


linear_est = tf.estimator.LinearClassifier(feature_columns=fc) #This is basically our ML model
linear_est.train(train_input_fn)  # train
result = linear_est.evaluate(eval_input_fn)  # get model metrics/stats by testing on tetsing data

clear_output()  # clears consoke output
#print(result)  # the result variable is simply a dict of stats about our model

res = list(linear_est.predict(eval_input_fn))      #Linear_est is our trained model. 
print(y_eval.loc[20])
print(res[67]['probabilities'][0])                     #res[0] is basically the prediction for the first person on the dataset.


```
